```python
import pandas as pd
url = 'https://github.com/owid/covid-19-data/raw/master/public/data/owid-covid-data.csv'
df = pd.read_csv(url, index_col=0)
print(df)
```

```python
#df.info()
dick = pd.DataFrame(df)
dick
```

```python
kk=dick[['location' , 'date', 'new_cases']] #選出需要的部分
#dick['Result'] = dick['new_cases']/dick['total_cases']
#dick[['location' , 'Result']]
```

```python
#df.drop(df.loc[df['location']=='World'].index, inplace=True)
max_cases = kk.sort_values(by=['new_cases'], ascending=False) #對確診資料由高到低做排序
max_cases #第一大題第一小題解答～找出資料中全球確診人數最高的一天-2021-01-07!
```

```python
aa=kk[(kk['date'] >= '2020-12-01') & (df['date'] <= '2021-12-31')]
bb=kk[(kk['date'] >= '2020-11-01') & (df['date'] <= '2021-11-31')]#選出一段日期
#kk.iloc[281:311]
```

```python
aa #十二月份的確診數
```

```python
bb #十一月份的確診數
```

```python
twsum = aa.groupby('location').sum() #依照國家 group 起來並做十二月份的加總
final_twsum = twsum.drop(index='World')
final_twsum

#Find a duplicate rows
#new_b = df.drop_duplicates()
#new_b
#d=new_b.values.reshape(1,193)
#Total = kk['new_cases'].sum()
#aa.drop(aa.loc[df['location']=='World'].index, inplace=True)
#final_twsum[['Hong Kong']]
```

```python
elsum = bb.groupby('location').sum() #依照國家 group 起來並做十一月份的加總
final_elsum = elsum.drop(index='World') #剔掉全球的部分
final_elsum 
```

```python
son = final_twsum - final_elsum
son #找到分子的部分（當月份減掉前月份）
```

```python
son.values.reshape(12,16) #把 array reshape 看一下全部國家的狀態
```

```python
growth_rate = son/final_twsum #成長率=當月份減掉前月份/當月份確診人數
growth_rate 
```

```python
oh = growth_rate.sort_values(by=['new_cases'])
oh #第一大題第二小題解答～找出確診國家中，12月新增確診”成長率"最高的三個國家 
```
