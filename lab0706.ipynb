# Dealing with med dataset

```python
dfnew_15 = df15.drop(['ChartNoId','藥名','劑量','頻次','服法','用藥時間'], axis = 1)
dfnew_15
```

```python
#create med_dictionary
med_dict = dict.fromkeys(['ZOCOT','LESC8','LIPI2','CREST','LIVAT','LINIT','VYTOT','CADUT'],'STAINS')
med_dict.update(dict.fromkeys(['CAPOT','ZESTR','TRIT1','MONOP','ACERT','ACERP'],'ACEIS'))
med_dict.update(dict.fromkeys(['INDE1','INDE4','BETAZ','TENOR','CONC1','CONCT'],'B-blockers'))
med_dict.update(dict.fromkeys(['LOSEI','PANTI','PANTT','TAKET','TAKE3','PARIT','NEXIT','NEXII','DEXIT'],'PPI'))
med_dict.update(dict.fromkeys(['CMT3','ZANTA','ZANTI','GASTE','GASTI','ROXAT'],'H2'))
med_dict.update(dict.fromkeys(['PLAVT','PLAV3'],'anticoagulants'))
med_dict.update(dict.fromkeys(['PLET1','PLETT'],'Cilostazol'))
med_dict.update(dict.fromkeys(['SCODO','PARML','PERIO'],'Oral care'))

#map 院內代碼_to_med_dictionary
dfnew_15['Medtype'] = dfnew_15['院內代碼'].map(med_dict)
dfnew_15['Medtype'] = dfnew_15['Medtype'].fillna('Others')
#one hot
dfoh_15 = pd.get_dummies(dfnew_15, prefix=['Medtype'], columns=['Medtype'])
dfoh_15
#取特定欄位
#dfnew_15 = dfoh_15.drop(['ChartNoId','院內代碼','藥名','劑量','頻次','服法','用藥時間'], axis = 1)
dfnew_15 = dfoh_15.drop(['院內代碼'], axis = 1)
#dfoh_15.columns


#https://www.dezyre.com/recipes/map-values-in-pandas-dataframe
#https://stackoverflow.com/questions/2974022/is-it-possible-to-assign-the-same-value-to-multiple-keys-in-a-dict-object-at-onc
#https://stackoverflow.com/questions/37265312/how-to-create-dummies-for-certain-columns-with-pandas-get-dummies
```

```python
#將相同病例號重複資料刪除，並利用 row 相加合併
df_without_duplicates = dfnew_15.drop_duplicates()
df_sum = df_without_duplicates.groupby('ChartNo', as_index=False).sum()
df_sum
```

```python
#變更欄位名稱
df15_0707_new = df15_0707.rename(columns={'住院序號':'ChartNoId','病歷':'ChartNo','處置代碼':'院內代碼','處置名稱':'藥名'})
df15_0707_new.columns
```

```python
dfnew_df15_0707 = df15_0707_new.drop(['ChartNoId','藥名'], axis = 1)
dfnew_df15_0707
```

```python
#map 院內代碼_to_med_dictionary
dfnew_df15_0707['Medtype'] = dfnew_df15_0707['院內代碼'].map(med_dict)
dfnew_df15_0707['Medtype'] = dfnew_df15_0707['Medtype'].fillna('Others')
#one hot
dfohshit_15 = pd.get_dummies(dfnew_df15_0707, prefix=['Medtype'], columns=['Medtype'])
dfohshit_15
#取特定欄位
dfnew_15_0707 = dfohshit_15.drop(['院內代碼'], axis = 1)
#將相同病例號重複資料刪除，並利用 row 相加合併
dfnew_15_0707_without_duplicates = dfnew_15_0707.drop_duplicates()
df_0707_sum = dfnew_15_0707_without_duplicates.groupby('ChartNo', as_index=False).sum()
df_0707_sum
```

```python
#合併新舊藥物分類資料
medwhole = pd.concat([df_0707_sum, df_sum], ignore_index=True)

#檢查新舊藥物資料是否有重複病歷號
medwhole_without_duplicates = medwhole.drop_duplicates()

#將相同病例號重複資料刪除，並利用 row 相加合併
med_whole = medwhole_without_duplicates.groupby('ChartNo', as_index=False).sum()
#medwhole -> 2822+920 = 共3742筆
#medwhole_without_duplicates共3742筆(新舊資料無重複病歷號)
med_whole
```
